```python
import numpy as np
import scipy.stats as stats
# Create a dummy dataset of 10 year old children’s weight
user_input=input("Enter weights of 10 year old children separated by commas:")
data= np.array([float(x) for x in user_input.split(",")]) 
# Define the null hypothesis H0 = ”The average weight of 10 year old children is 32kg.”
# Define the alternative hypothesis H1 = ”The average weight of 10 year old children is more than 32kg.
# Calculate the test statistic
t_stat, p_value = stats.ttest_1samp(data, 32)
# Print the results
print("Test statistic:", t_stat)
print("p-value:", p_value)
# Conclusion
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
import matplotlib.pyplot as plt #for core plotting
import seaborn as sns #for beautiful and statistical plots

sns.histplot(data, kde=True) #for creating smoothed version of histograms.
plt.axvline(np.mean(data), color='r', linestyle='--', label='Mean')
plt.axvline(32, color='g', linestyle='--', label='Hypothesized Mean')
plt.legend()
plt.title('Distribution of Weights')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show() #display plot
```
